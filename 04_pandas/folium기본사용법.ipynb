```python
m = folium.Map([36.6108968,127.2864676],  zoom_start=15)

icon = folium.Icon( color='red',
                                        icon_color='white',
                                        icon='home',
                                        angle=0)
folium.Marker([36.6108968,127.2864676],
                                tooltip="<b>Click me!!</b><br>툴팁사용",
                                 popup="<b>눌러주세요!</b><br>팝업사용",
                                 icon=icon).add_to(m)
m
```
