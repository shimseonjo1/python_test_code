```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(-100,100,50).cumsum()
data
```

```python
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```

```python
# 1. 폰트 파일을 FontProperties()생성하고 사용
```

```python
import matplotlib.font_manager as fm
path ='NanumBarunGothic.ttf'
fontprop = fm.FontProperties(fname=path,size=12)
data = np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이',fontproperties=fontprop)
plt.ylabel('주식가격',fontproperties=fontprop)
plt.xlabel('시간(분)',fontproperties=fontprop)
```

```python
# 2. matplotlib.rcParams[]로 설정
```

```python
import matplotlib
print('설정파일위치:',matplotlib.matplotlib_fname())
```

```python
[(file.name,file.fname) for file in fm.fontManager.ttflist if 'Nanum' in file.name]
```

```python
plt.rcParams['font.family'] = 'Nanum Pen Script'
plt.rcParams['axes.unicode_minus']=False
```

```python
data = np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```
